```python
# Load tokenizer and model
model_name = "jbeno/electra-base-classifier-sentiment"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = ElectraClassifier.from_pretrained(model_name)

# Set model to evaluation mode
model.eval()
```

```python
from google.colab import files
import pandas as pd
import io

# Step 3: Upload your CSV file (should have a 'text' column)
uploaded = files.upload()
filename = next(iter(uploaded))  # Get uploaded filename

# Step 5: Load the CSV
# if its csv file
# remove this comment and comment the later part
#df = pd.read_csv(io.BytesIO(uploaded[filename]))

# if its file of sentences
with open(filename, 'r', encoding='utf-8') as f:
    sentences = [line.strip() for line in f if line.strip()]

```

IF FILE IS COLLECTION OF SENTENCES

```python
sentiments = []
for i, text in enumerate(sentences):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=512)
    with torch.no_grad():
        logits = model(**inputs)
        predicted_class_id = torch.argmax(logits, dim=1).item()
        predicted_label = model.config.id2label[predicted_class_id]
        sentiments.append(predicted_label)

    # Print first 10 outputs
    if i < 10:
        print(f"\n🔹 Sentence {i+1}: {text}\n   → Sentiment: {predicted_label}")
```

For such file if new csv is to be downloaded

```python
# --- STEP 4: CREATE CSV OUTPUT ---
df = pd.DataFrame({
    'sentence': sentences,
    'sentiment': sentiments
})

output_file = "sentiment_results2.csv"
df.to_csv(output_file, index=False)

# --- STEP 5: DOWNLOAD RESULT ---
print("\n Downloading full results as CSV...")
files.download(output_file)
```
