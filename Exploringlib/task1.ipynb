```python
from google.colab import files
upload= files.upload()

for name in upload.keys():
    filename = name
```

```python
with open(filename, "rt") as f:
  text=f.read()
```

```python
import nltk
nltk.download('punkt_tab')

from nltk.tokenize import sent_tokenize, word_tokenize

sentences=sent_tokenize(text)
words=word_tokenize(text)

print("Total no.of sentence", len(sentences))
print("Total no.of words", len(words))

print("\nSample sentences:")
for s in sentences[:3]:
    print("-", s)

print("\nSample words:", words[:10])
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))
```

To remove the unwanted punctuation marks


```python
import re

punctuation = re.compile(r'[^a-zA-Z]+')  # keep only letters

# Clean each word
post= []
for word in words:
    cleaned = punctuation.sub("", word)
    if len(cleaned) > 0:
        lower_word = cleaned.lower()
        if lower_word not in stop_words:
            post.append(lower_word)
```

```python
from nltk.util import ngrams

gram1=list(nltk.ngrams(post,2))
gram2=list(nltk.ngrams(post,3))
# using text here instead of words gives o/p wrt each character

print(gram1)
print(gram2)
```

```python
from collections import Counter

freq1 = Counter(gram1)
freq2 = Counter(gram2)

print("2-gram frequencies:")
for phrase, freq in freq1.items():
    print(f"{phrase}: {freq}")

print("\n3-gram frequencies:")
for phrase, freq in freq2.items():
    print(f"{phrase}: {freq}")
```

```python
# same thing only but trying with differebt function


from nltk import FreqDist

freq_bigrams = FreqDist(gram1)
freq_trigrams = FreqDist(gram2)

# Print frequencies
print("Bigram frequencies:")
for phrase, freq in freq_bigrams.items():
    print(f"{phrase}: {freq}")

print("\nTrigram frequencies:")
for phrase, freq in freq_trigrams.items():
    print(f"{phrase}: {freq}")
```
